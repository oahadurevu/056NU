```python
import math
def func(x):
    return (1/(1+x*x))

def Simpsons_Rule(x0, xn, n):
    h = (xn-x0)/n
    total = func(x0) + func(xn)
    xi = 0.0
    for i in range (1,n):
        xi = x0 + h
        if (i%2 == 0):
            total += 2*func(xi)
        else:
            total += 4*func(xi)
        x0 = xi

    return (h/3) * total

x0  = float(input("Enter the lower limit: "))
xn  = float(input("Enter the upper limit: "))
n  = int(input("Enter the number of interval: "))

approx_area = Simpsons_Rule(x0, xn, n)
print("=> By using Simpsons method: ")
print(f"The obtained approximate area is: {approx_area:.4f}")

exact_sol = math.atan(xn) - math.atan(x0)
print(f"The actual area is: {exact_sol:.4f}")
error = abs(exact_sol - approx_area)
print(f"The error is: {error:.4f}")
```

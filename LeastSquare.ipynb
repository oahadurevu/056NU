```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0,2,5,7])
y = np.array([-1,5,12,20])
m = len(x)

sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xy = np.sum(x * y)
sum_x2 = np.sum(x * x)

x_mean = sum_x / m
y_mean = sum_y / m

a1 = ((m * sum_xy) - (sum_x * sum_y)) / ((m * sum_x2) - (sum_x ** 2))
a0 = y_mean - (a1 * x_mean)

print(f"slope(a1): {a1:.2f}")
print(f"intercept(a0): {a0:.2f}")

y_pred = a0 + a1 * x

plt.scatter(x, y, color="red", label="data points")
plt.plot(x, y_pred, color="blue", label="Best fit Line")

plt.legend()
plt.show()
```
